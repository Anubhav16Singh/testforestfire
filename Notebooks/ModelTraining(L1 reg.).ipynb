```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('Algerian_forest_fires_cleaned_dataset.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df.drop(['day', 'month', 'year'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Classes'].value_counts()
```

```python
df['Classes'] = df['Classes'].astype(str)


df['Classes'] = np.where(df['Classes'].str.contains('not fire'), 0, 1)
```

```python
df['Classes'].value_counts()
```

```python
##Independent and depemdent features
X=df.drop(['FWI'],axis=1)
y=df['FWI']
```

```python
y
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
X_train.shape
```

```python
##Feature selection based on correlation
```

```python
X_train.corr()
```

```python
##Check for multiCollinearity
plt.figure(figsize=(12,10))
corr=X_train.corr()
sns.heatmap(corr,annot=True)
```

```python
def correlation(dataset,threshold):
    col_corr=set()
    corr_matrix=dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j])>threshold:
                colname=corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr           
```

```python
corr_features=correlation(X_train,0.85)
```

```python
##Dropfeatures when correlation is more than 0.85
X_train.drop(corr_features,axis=1,inplace=True)
X_test.drop(corr_features,axis=1,inplace=True)
X_train.shape
```

```python
##Feature Scaling 
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
X_test_scaled
```

```python
##Box plots
```

```python
plt.subplots(figsize=(15,5))
plt.subplot(1,2,1)
sns.boxplot(data=X_train)
plt.title('X_train before Scaling')
plt.subplot(1,2,2)
sns.boxplot(data=X_test_scaled)
plt.title('X_train after scaling')
```

```python
#Cross validation
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,r2_score
linereg=LinearRegression()
linereg.fit(X_train_scaled,y_train)
y_pred=linereg.predict(X_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean absolute error:",mae)
print("R2  score:",score)
plt.scatter(y_test,y_pred)
```

```python
##Lasso Regression
```

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_absolute_error,r2_score
lasso=Lasso()
lasso.fit(X_train_scaled,y_train)
y_pred=lasso.predict(X_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean absolute error:",mae)
print("R2  score:",score)
plt.scatter(y_test,y_pred)
```

```python
##Cross validation Lasso Regression

from sklearn.linear_model import LassoCV
lassocv=LassoCV(cv=5)
lassocv.fit(X_train_scaled,y_train)
ypred=lassocv.predict(X_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean absolute error:",mae)
print("R2 score:",score)
plt.scatter(y_test,y_pred)
```
